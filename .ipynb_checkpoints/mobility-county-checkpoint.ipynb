```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Global_Mobility_Report.csv",low_memory=False)
data=data.rename(columns = {
        "country_region":"country",
        "sub_region_1":"state",
        "sub_region_2":"county",
    "retail_and_recreation_percent_change_from_baseline":"retail",
"grocery_and_pharmacy_percent_change_from_baseline":"grocery_pharmacy",
    "parks_percent_change_from_baseline":"parks",
    "transit_stations_percent_change_from_baseline":"transit",
    "workplaces_percent_change_from_baseline":"transit",
"residential_percent_change_from_baseline":"residential"})
```

```python
data=data.dropna(subset =['retail','grocery_pharmacy','parks','transit','transit','residential']).drop(columns=['census_fips_code'])
data
    
```

```python
means=data.mean()
```

```python
g= data.groupby( ['country','state', 'county'])
```

```python
g.mean().sort_values(['retail'],ascending=False )
```

```python
norms =g.mean()-means
norms
```

```python
norms.sort_values(['retail'],ascending=False )
```
